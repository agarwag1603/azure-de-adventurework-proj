```python
%python
tenant_id = ""
client_id = ""
client_secret = ""
storage_account_name = "dataengineeringstore1603"

configs = {
    "fs.azure.account.auth.type": "OAuth",
    "fs.azure.account.oauth.provider.type": "org.apache.hadoop.fs.azurebfs.oauth2.ClientCredsTokenProvider",
    "fs.azure.account.oauth2.client.id": client_id,
    "fs.azure.account.oauth2.client.secret": client_secret,
    "fs.azure.account.oauth2.client.endpoint": f"https://login.microsoftonline.com/{tenant_id}/oauth2/token"
}

container_name = "adventure-work-container"
mount_point = "/mnt/adventurework"

dbutils.fs.mount(
    source=f"abfss://{container_name}@{storage_account_name}.dfs.core.windows.net/",
    mount_point=mount_point,
    extra_configs=configs
)

# Verify the connection
display(dbutils.fs.ls(mount_point))
```

```python
AdventureWorks_Calendar = spark.read.format("csv")\
                            .load(f"{mount_point}/AdventureWorks_Calendar/AdventureWorks_Calendar.csv", header=True, inferSchema=True)
```

```python
%python
mount_point = "/mnt/adventurework"

# Get the list of files and directories in the mount point
file_list = dbutils.fs.ls(mount_point)

# Dictionary to store DataFrames
dataframes = {}

# Loop through the file list
for file in file_list:
    # Access the file properties
    file_path = file.path
    file_name = file.name


    # Load the CSV file into a DataFrame
    dataframe = spark.read.format("csv").load(
        file_path,
        header=True,
        inferSchema=True
    )

    # Store the DataFrame in the dictionary
    dataframes[file_name] = dataframe

    # Perform any desired operations with the file properties
    print(f"File path: {file_path}")
    print(f"File name: {file_name}")


# Example: Access a specific DataFrame
# display(dataframes["AdventureWorks_Calendar.csv"])
```

```python
dataframes['AdventureWorks_Calendar/']
```

```python
from pyspark.sql.functions import month, year,dayofmonth

adventureworks_calendar_df = dataframes['AdventureWorks_Calendar/']


adventureworks_calendar_df=adventureworks_calendar_df.withColumn('Month', month(adventureworks_calendar_df['Date']))
adventureworks_calendar_df=adventureworks_calendar_df.withColumn('Year', year(adventureworks_calendar_df['Date']))
adventureworks_calendar_df=adventureworks_calendar_df.withColumn('Day', dayofmonth(adventureworks_calendar_df['Date']))

```

```python
adventureworks_calendar_df.display()
```

```python
adventureworks_calendar_df.write.mode("append").option("header","true").parquet("/mnt/adventurework/AdventureWorks_Calendar/adventureworks_calendar_df.parquet")
```
